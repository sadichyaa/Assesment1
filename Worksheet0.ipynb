## **4.1 Exercise on Functions:**
Task - 1:

Create a Python program that converts between different units of measurement.

• The program should:
1. Prompt the user to choose the type of conversion (e.g., length, weight, volume).
2. Ask the user to input the value to be converted.
3. Perform the conversion and display the result.
4. Handle potential errors, such as invalid input or unsupported conversion types.

• Requirements:

1. Functions: Define at least one function to perform the conversion.
2. Error Handling: Use try-except blocks to handle invalid input (e.g., non-numeric values).
3. User Input: Prompt the user to select the conversion type and input the value.
4. Docstrings: Include a docstring in your function to describe its purpose, parameters, and
return value.

• Conversion Options:

1. Length:

– Convert meters (m) to feet (ft).

– Convert feet (ft) to meters (m).

2. Weight:

– Convert kilograms (kg) to pounds (lbs).

– Convert pounds (lbs) to kilograms (kg).

3. Volume:

– Convert liters (L) to gallons (gal).

– Convert gallons (gal) to liters (L).


```python
def convert_units(conversion_type, value):
    """
    Convert units based on the selected conversion type.

    Parameters:
    conversion_type (int): The conversion option chosen by the user.
    value (float): The numeric value to convert.

    Returns:
    float: The converted value.
    """

    if conversion_type == 1:  # meters to feet
        return value * 3.28084
    elif conversion_type == 2:  # feet to meters
        return value / 3.28084
    elif conversion_type == 3:  # kg to pounds
        return value * 2.20462
    elif conversion_type == 4:  # pounds to kg
        return value / 2.20462
    elif conversion_type == 5:  # liters to gallons
        return value * 0.264172
    elif conversion_type == 6:  # gallons to liters
        return value / 0.264172
    else:
        return None


def main():
    print("Unit Conversion Program")
    print("1. Meters to Feet")
    print("2. Feet to Meters")
    print("3. Kilograms to Pounds")
    print("4. Pounds to Kilograms")
    print("5. Liters to Gallons")
    print("6. Gallons to Liters")

    try:
        choice = int(input("Choose conversion type (1-6): "))
        value = float(input("Enter the value to convert: "))

        result = convert_units(choice, value)

        if result is None:
            print("Unsupported conversion type.")
        else:
            print("Converted value:", result)

    except ValueError:
        print("Invalid input! Please enter numeric values.")


main()
```

## **Task - 2:**
Create a Python program that performs various mathematical operations on a list of numbers.

• The Program should:

1. Prompt the user to choose an operation (e.g., find the sum, average, maximum, or minimum
of the numbers).

2. Ask the user to input a list of numbers (separated by spaces).

3. Perform the selected operation and display the result.

4. Handle potential errors, such as invalid input or empty lists.

• Requirements:

1. Functions: Define at least one function for each operation (sum, average, maximum, mini-
mum).

2. Error Handling: Use try-except blocks to handle invalid input (e.g., non-numeric values or
empty lists).

3. User Input: Prompt the user to select the operation and input the list of numbers.

4. Docstrings: Include a docstring in each function to describe its purpose, parameters, and
return value.

```python
def find_sum(numbers):
    """
    Calculate the sum of a list of numbers.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: The sum of the numbers.
    """
    return sum(numbers)


def find_average(numbers):
    """
    Calculate the average of a list of numbers.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: The average value.
    """
    return sum(numbers) / len(numbers)


def find_maximum(numbers):
    """
    Find the maximum value in a list of numbers.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: The maximum value.
    """
    return max(numbers)


def find_minimum(numbers):
    """
    Find the minimum value in a list of numbers.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: The minimum value.
    """
    return min(numbers)


def main():
    print("Mathematical Operations on a List")
    print("1. Find Sum")
    print("2. Find Average")
    print("3. Find Maximum")
    print("4. Find Minimum")

    try:
        choice = int(input("Choose an operation (1-4): "))
        user_input = input("Enter numbers separated by spaces: ")

        numbers = [float(num) for num in user_input.split()]

        if len(numbers) == 0:
            print("Error: The list cannot be empty.")
            return

        if choice == 1:
            print("Sum:", find_sum(numbers))
        elif choice == 2:
            print("Average:", find_average(numbers))
        elif choice == 3:
            print("Maximum:", find_maximum(numbers))
        elif choice == 4:
            print("Minimum:", find_minimum(numbers))
        else:
            print("Invalid operation choice.")

    except ValueError:
        print("Invalid input! Please enter numeric values only.")


main()
```

```python
# 1. Extract Every Other Element
def extract_every_other(lst):
    """
    Return every other element from the list starting from the first element.

    Parameters:
    lst (list): Input list.

    Returns:
    list: List containing every other element.
    """
    return lst[::2]


# 2. Slice a Sublist
def get_sublist(lst, start, end):
    """
    Return a sublist from start index to end index (inclusive).

    Parameters:
    lst (list): Input list
    start (int): Starting index
    end (int): Ending index

    Returns:
    list: Sublist from start to end
    """
    return lst[start:end+1]


# 3. Reverse a List Using Slicing
def reverse_list(lst):
    """
    Return the reversed version of the list using slicing.

    Parameters:
    lst (list): Input list

    Returns:
    list: Reversed list
    """
    return lst[::-1]


# 4. Remove the First and Last Elements
def remove_first_last(lst):
    """
    Remove the first and last elements from a list.

    Parameters:
    lst (list): Input list

    Returns:
    list: List without first and last elements
    """
    return lst[1:-1]


# 5. Get the First n Elements
def get_first_n(lst, n):
    """
    Return the first n elements of the list.

    Parameters:
    lst (list): Input list
    n (int): Number of elements to extract

    Returns:
    list: First n elements
    """
    return lst[:n]


# 6. Extract Last n Elements
def get_last_n(lst, n):
    """
    Return the last n elements of the list.

    Parameters:
    lst (list): Input list
    n (int): Number of elements

    Returns:
    list: Last n elements
    """
    return lst[-n:]


# 7. Extract Elements in Reverse Order Skipping One
def reverse_skip(lst):
    """
    Return every second element starting from the second-to-last
    element while moving backward.

    Parameters:
    lst (list): Input list

    Returns:
    list: Elements extracted in reverse with skipping
    """
    return lst[-2::-2]


# Example usage
numbers = [1, 2, 3, 4, 5, 6]

print("Every other:", extract_every_other(numbers))
print("Sublist:", get_sublist(numbers, 2, 4))
print("Reversed:", reverse_list(numbers))
print("Remove first & last:", remove_first_last(numbers))
print("First 3:", get_first_n(numbers, 3))
print("Last 2:", get_last_n(numbers, 2))
print("Reverse skip:", reverse_skip(numbers))
```

8. Nested List Average:

Write a Python function that calculates the average of all elements in a nested list.

• Requirements:

– Define a function average nested(lst) that takes a nested list lst and returns the average
of all the elements.

– Example: For the input [[1, 2], [3, 4], [5, 6]], the output should be 3.5.

```python
# 1. Flatten a Nested List
def flatten(lst):
    """
    Flatten a nested list into a single list (one level).

    Parameters:
    lst (list): A nested list.

    Returns:
    list: Flattened list.
    """
    result = []
    for sub in lst:
        result.extend(sub)
    return result


# 2. Access Nested Element
def access_nested_element(lst, indices):
    """
    Access an element from a nested list using a list of indices.

    Parameters:
    lst (list): Nested list
    indices (list): List of indices

    Returns:
    element: Element at the given position
    """
    element = lst
    for i in indices:
        element = element[i]
    return element


# 3. Sum of All Elements in a Nested List
def sum_nested(lst):
    """
    Calculate the sum of all numbers in a nested list.

    Parameters:
    lst (list): Nested list

    Returns:
    int/float: Sum of all elements
    """
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total


# 4. Remove Specific Element
def remove_element(lst, elem):
    """
    Remove all occurrences of elem from a nested list.

    Parameters:
    lst (list): Nested list
    elem (value): Element to remove

    Returns:
    list: Modified list
    """
    result = []
    for item in lst:
        if isinstance(item, list):
            result.append(remove_element(item, elem))
        elif item != elem:
            result.append(item)
    return result


# 5. Find Maximum Element
def find_max(lst):
    """
    Find the maximum element in a nested list.

    Parameters:
    lst (list): Nested list

    Returns:
    number: Maximum value
    """
    max_val = float('-inf')
    for item in lst:
        if isinstance(item, list):
            max_val = max(max_val, find_max(item))
        else:
            max_val = max(max_val, item)
    return max_val


# 6. Count Occurrences
def count_occurrences(lst, elem):
    """
    Count occurrences of elem in a nested list.

    Parameters:
    lst (list): Nested list
    elem (value): Element to count

    Returns:
    int: Number of occurrences
    """
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        elif item == elem:
            count += 1
    return count


# 7. Deep Flatten
def deep_flatten(lst):
    """
    Flatten a deeply nested list into a single list.

    Parameters:
    lst (list): Deeply nested list

    Returns:
    list: Completely flattened list
    """
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(deep_flatten(item))
        else:
            result.append(item)
    return result


# 8. Average of Nested List
def average_nested(lst):
    """
    Calculate the average of all elements in a nested list.

    Parameters:
    lst (list): Nested list

    Returns:
    float: Average value
    """
    flat = deep_flatten(lst)
    return sum(flat) / len(flat)


# Example Usage
data = [[1, 2], [3, [4, 5]], 6]

print("Flatten:", flatten([[1,2],[3,4],[5]]))
print("Access element:", access_nested_element([[1,2,3],[4,5,6],[7,8,9]],[1,2]))
print("Sum nested:", sum_nested(data))
print("Remove element:", remove_element([[1,2],[3,2],[4,5]],2))
print("Max element:", find_max(data))
print("Count occurrences:", count_occurrences([[1,2],[2,3],[2,4]],2))
print("Deep flatten:", deep_flatten([[[1,2],[3,4]],[[5,6],[7,8]]]))
print("Average:", average_nested([[1,2],[3,4],[5,6]]))
```

```python
import numpy as np
import time
import random

# size for vector operations
size = 1000000

# create python lists
list1 = [random.random() for _ in range(size)]
list2 = [random.random() for _ in range(size)]

# create numpy arrays
arr1 = np.array(list1)
arr2 = np.array(list2)


# 1. Element-wise Addition


# Python list addition
start = time.time()
result = [list1[i] + list2[i] for i in range(size)]
end = time.time()
print("Python list addition time:", end - start)

# NumPy addition
start = time.time()
result = arr1 + arr2
end = time.time()
print("NumPy addition time:", end - start)


# 2. Element-wise Multiplication


# Python list multiplication
start = time.time()
result = [list1[i] * list2[i] for i in range(size)]
end = time.time()
print("Python list multiplication time:", end - start)

# NumPy multiplication
start = time.time()
result = arr1 * arr2
end = time.time()
print("NumPy multiplication time:", end - start)


# 3. Dot Product


# Python list dot product
start = time.time()
dot = sum(list1[i] * list2[i] for i in range(size))
end = time.time()
print("Python list dot product time:", end - start)

# NumPy dot product
start = time.time()
dot = np.dot(arr1, arr2)
end = time.time()
print("NumPy dot product time:", end - start)


# 4. Matrix Multiplication


# smaller matrices recommended for Python lists
n = 200

A = [[random.random() for _ in range(n)] for _ in range(n)]
B = [[random.random() for _ in range(n)] for _ in range(n)]

# Python list matrix multiplication
start = time.time()

result = [[0]*n for _ in range(n)]

for i in range(n):
    for j in range(n):
        for k in range(n):
            result[i][j] += A[i][k] * B[k][j]

end = time.time()
print("Python matrix multiplication time:", end - start)

# NumPy matrices
A_np = np.array(A)
B_np = np.array(B)

start = time.time()
result = np.dot(A_np, B_np)
end = time.time()
print("NumPy matrix multiplication time:", end - start)
```
